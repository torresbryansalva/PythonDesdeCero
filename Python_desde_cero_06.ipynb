## FUNCIONES EN PYTHON


```python
#funcion basica
def fun():
  return 3
print("valor de la función:",fun())
```

```python
#funcion con argumentos
def fun1(x):
  return 2*x-3
y=fun1(2)
y
```

```python
#varios argumentos
def fun2(x,y):
  z=x+y
  return z
y=fun2(2,4)
y
```

```python
#parametros por defecto
def fun3(a=1,b=3,c=-3):
  z=a+b+c
  return z
y=fun3()
y
```

```python
def fun4(num):
  x=0
  for i in num:
    x=x+i
  return x
y=fun4([1,2,3,4,5])
y
```

```python
def fun5(un):
  x=0
  for i in range(1,un):
    x+=i
  return x
y=fun5(5)
y
```

```python
#diccionario como parametro
def fun6(**x):
  suma=0
  for i,j in x.items():
    suma=suma+j
  return suma

dic={"a":2,"b":4,"c":-5,"d":12}
y=fun6(**dic)
y
```

```python
#devolviendo varios documentos
def fun7(x,y):
  suma=x+y
  resta=x-y
  return suma,resta
i,j=fun7(4,5)
print(i)
print(j)
```

## PASO POR VALOR Y REFERENCIA
* Si usamos un parámetro pasado por **valor**, se creará una copia local de la variable, lo que implica que cualquier modificación sobre la misma no tendrá efecto sobre la original.
* Con una variable pasada como **referencia**, se actuará directamente sobre la variable pasada, por lo que las modificaciones afectarán a la variable original.

```python
#paso por valor
x=1
def func(entrada):
  entrada=0

y=func(x)
print("paso por valor",x)
```

```python
#paso por referencia
xm=[1,2,3]
def func1(entrada):
    entrada.append(40)

func1(xm)
print("paso por referencia",xm)
```

```python
#no se destruye la lista orgianal
re=[1,2,3,4,5]
def func2(argumento):
  argumento=[]
func2(re)
print(re)
```

## ARGUMENTOS



```python
def funn(a,b,c):
  s=a+b+c
  return s
print(funn(1,2,3))
```

```python
#varios argumentos con *arg no es puntero es una forma de decir que hay varios argumentos
def funa(*args):
  s=0
  for i in args:
    s=s+i
  return s

print(funa(1,2,3,4,5,5))
print(funa(1,2))
```

```python
def funcc(**kwargs):
  for i,j in kwargs.items():
    print(i,"=",j)
funcc(a=1,b=4,c=12,d=2)
```

```python
#combinando todos los tipos
def fung(a,b,*args,**kwargs):
  print("a=",a)
  print("b=",b)
  for i in args:
    print("args=",i)
  for j,k in kwargs.items():
    print(j,"=",k)

a=1
b=2
ls=[12,11,10]
dic={"m":12,"n":2,"p":-2}
print(fung(a,b,*ls,**dic))# ojo no olvidar poner el asterisco 
```

## LA RECURSIVIDAD

```python
def funaa(x):
  r=1
  n=1
  while r<=x:
    n=n*r
    r+=1
  return n

print(funaa(5))
```

```python
#en forma recursiva del FACTORIAL
def recursivo(x):
  if x==1:
    return 1
  else:
    return x*recursivo(x-1)

print(recursivo(4))
```

```python
#seria de fibonacci en forma recursiva
def fibonacci(x):
  if x==0:
    return 0
  elif x==1:
    return 1
  else:
    return fibonacci(x-1) + fibonacci(x-2)
x=10    
print("Numero en la posicion",x,"en la serie es:",fibonacci(x))
```

# Programacion Funcional

```python
lista = [1, 2, 3, 4, 5]
nlista = []
for l in lista:
    nlista.append(l*2)

print(nlista)
```

```python
def funpor(x):
  return x*2

nlista=map(funpor,lista)
print(list(nlista))
nlista
```

```python
#compactando un poco más
nlista=map(lambda x:x*2,lista)
print(list(nlista))
```

### fitrar

```python
lista = [1,10,12,7, 4, 16, 3, 8]
pares = filter(lambda x: x % 2 == 0, lista)

print(list(pares))
```

```python
lista = [1, 0, 1, 1, 0]
pares = filter(lambda x: x == 0, lista)

print(list(pares))
```
