## DICCIONARIOS

son una forma de estructura en donde podemos almacenar un **identificador** con su **valor** de la forma **key:valor**

```python
dic={
    "Nombre":"Cedro",
    "Ciudad":"Chachapoyas",
    "Año":23,
}
print(dic)
```

```python
#otra forma es 
dict1=dict([
           ("Nombre","Pedro"),
           ("Apellido","Salva"),
           ("edad",233)
])
print(dict1)
```

```python
#la mejor forma creo yo
dict2=dict(
    Nombre="Bryan",
    Apellido="Torres",
    Altura=177
)
print(dict2)
```

#ALGUNAS PROPIEDADES
* SON DINAMICOS: Es decir que puden crecer o decrecer 
* SON INDEXADOS: Es decir se puede acceder a sus valores a traves de su indice
* SON ANIDADOS: Pueden almacenar a otro diccionario en su parametro **valor**

```python
#accediendo a sus elementos
print(dict1['Nombre'])
print(dict1.get('Nombre'))
```

```python
#modificar un elemento
dict2["Nombre"]="Luciana"
print(dict2)
```

```python
#creacion de otras variables
dict2["Peso"]="345 kg"
print(dict2)
```

```python
#ejemplo1 imprimiendo los keys
for i in dict2:
  print(i)
```

```python
#imprimiendo los valores
for i in dict2:
  print(dict2[i])
```

```python
#imprimiendo ambos valores
for i,j in dict2.items():
  print(i,j)
```

```python
#diccionarios anidados
d1={"a":2,"b":3}
d2=dict(c=2,d="morvius")
d={"diccionario 1":d1,"diccionario 2":d2}
print(d)
```

```python
#OPERACIONES
d.clear()
print(d)
```
