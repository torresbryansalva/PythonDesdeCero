## SETS
No tienen un orden de almacenamiento, por lo que no se podr√° acceder a travez de los indices.

```python
#primera forma de declarar los sets
m=set([1,2,3,4,4,5,6,7,7])
m
print(m,type(m))
```

```python
#segunda forma de declarar a los sets
m1={1,2,3,4,4,5,6,7,7}
m1
print(m1,type(m1))
```

```python
#OPERACIONES

for i in m:
  print(i)

print("numero de elementos",len(m))
```

```python
a=set([1,2,3,4])
b=set(["arbol","cedro","ishingo"])
c=(a|b)
print(c)
```
