## USO DEL BREAK: son para romper el ciclo y salir de la funcion

```python
#ejemplo 10
var="Mundo matematico"
for i in var:
  if i=='a':
    print("Se encontro la letra buscada",i)
    break
  print(i)
```

```python
# ejemplo 11
for i in range(0,3):
  for j in range(0,4):
    break
  print(i,j)
```

## funcion zip()


```python
#ejemplo 12
a=[1,2,3]
b=["uno","dos","tres"]
c=zip(a,b)

print(list(c))
```

```python
#uso con el for
for i,j in zip(a,b):
  print("el numero:",i," y letra :",j)
```

## Numerate

```python
lista=["arbol","Montaña","lagunas","codores"]
indice=0

for i in lista:
  print(indice,i)
  indice=indice+1
```

```python
#ejemplo 14 USO DE enumerate que es mas pulcro

for indice,i in enumerate(lista):
  print(indice,i)

```

```python
#ejemplo
lista=["mono","aguila","colibri","vicuña"]
es=list(enumerate(lista))
print(es)
```
